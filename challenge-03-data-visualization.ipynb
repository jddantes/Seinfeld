### Function Viz

#### Draw the line `y = 3x + 2` consisting of 20 points spaced 0.5 each.
Use `np.linspace` to sample points, generate a function to simulate the equation and use `plotly` to draw the points.

Try replacing `3` and `2` with: a\) larger positive numbers, b) negative numbers, c) smaller positive numbers. What do you observe?

````python
import numpy as np
import plotly.graph_objects as go

# Define the line equation y = 3x + 2
def line_equation(x, a=3, b=2):
    return a * x + b

# Generate x values spaced 0.5 apart
x_values = np.linspace(-5, 5, num=20)

# Calculate corresponding y values using the line equation
y_values = line_equation(x_values, a=0.1, b=2)

# Create a scatter plot
fig = go.Figure()

# Add the line
fig.add_trace(go.Scatter(x=x_values, y=y_values, mode='lines', name='y = 3x + 2'))

# Add the points
fig.add_trace(go.Scatter(x=x_values, y=y_values, mode='markers', name='Points'))

# Update layout
# fig.update_layout(title='Line y = 3x + 2 with sampled points', xaxis=dict(title='x'), yaxis=dict(title='y'), aspectratio=dict(x=1, y=1))
# fig.update_layout(title='Line y = 3x + 2 with sampled points',
#                   xaxis=dict(title='x', scaleanchor="y", scaleratio=1),
#                   yaxis=dict(title='y', scaleanchor="x", scaleratio=1),
#                   aspectratio=dict(x=1, y=1))
# Show the plot
fig.show()
````

Steeper for higher slope, direction is flipped if negative
